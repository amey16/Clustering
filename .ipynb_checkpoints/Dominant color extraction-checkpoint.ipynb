```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from sklearn.cluster import KMeans
```

```python
im = cv2.imread('exp.jpg')
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
print(im.shape)

plt.figure()
plt.imshow(im)
```

```python
k = 4
kmeans = KMeans(k, init = 'random')
```

```python
reshaped_im = im.reshape((-1,3))
reshaped_im.shape
```

```python
kmeans.fit(reshaped_im)
```

```python
dominant_colors = kmeans.cluster_centers_.astype('uint8')
dominant_colors
```

```python
np.unique(kmeans.labels_, return_counts=True)
```

```python
plt.figure()
for i,color in enumerate(dominant_colors):
    palette = np.zeros_like(im, dtype='uint8')
    palette[:,:,:] = color
    plt.subplot(1,k,i+1)
    plt.axis("off")
    plt.imshow(palette)
print(im.shape)
```

```python
new_img = np.zeros_like(reshaped_im, dtype='uint8')
print(new_img.shape)
```

```python
for ix in range(new_img.shape[0]):
    new_img[ix] = dominant_colors[kmeans.labels_[ix]]
    
new_img = new_img.reshape((im.shape))
plt.imshow(new_img)
plt.show()
```
